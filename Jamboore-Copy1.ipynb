````python
def cond_prob(i, j):
    Joint_vals_fitness = pd.crosstab(i, j, normalize=True)
    product = Joint_vals_fitness.sum(axis=0)
    cond_Prob = Joint_vals_fitness.divide(product, axis=1)
    return cond_Prob.plot.bar(figsize=(13, 5),
                              grid=True,
                              color=('chartreuse', 'black', 'r', 'm',
                                     'darkorange', 'dimgrey', 'c', 'yellow',
                                     'deeppink'))
````

````python
df['cgpa_catagory'] = pd.cut(df.cgpa,bins=[6,7,8,9,10],labels=['6-7','7-8','8-9','9-10'])
````

````python
cond_prob(df.university_rating,df.cgpa_catagory)
plt.title(
    'Probability of statement of purpouse catagory wise according to university rating',
    fontsize=16,
    fontname='Comic Sans MS')
plt.show()
````

````python
def adj_r2_score(xtest,ytest,r2):
    r2 = reg.score(xtest,ytest)
    adjr2_sq = 1 - (1- r2)*(len(ytest)-1)/(len(ytest)-xtest.shape[1]-1)
    return adjr2_sq

def adj_r2_score1(xtest,ytest,r2):
    r2 = reg1.score(xtest,ytest)
    adjr2_sq = 1 - (1- r2)*(len(ytest)-1)/(len(ytest)-xtest.shape[1]-1)
    return adjr2_sq
````
