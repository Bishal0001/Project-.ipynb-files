```python
def cond_prob(a):
    Joint_vals_fitness = pd.crosstab(df['Education'],df['Product'],normalize=True)
    product = Joint_vals_fitness.sum(axis=0)
    cond_Prob = Joint_vals_fitness.divide(product,axis=1)
    cond_Prob.plot.bar(figsize=(15,6),grid=True,color={"black","m","grey"})
    plt.title('Every probability in this table represents a conditional probability of product given Education ie P( Product | Education ).')
    k = plt.show()
    return k

#cond_prob(a)


def remove_outlier(data):

    ans = []
    data_s = sorted(data)

    q1,q3 = np.percentile(data_s,[25,75])
    iqr = q1 - q3
    upper_bound = q3 - (1.5*iqr)
    lower_bound = q1 + (1.5*iqr)

    for i in range(len(data)):
        
        if data[i] > lower_bound and data[i] < upper_bound:
            ans.append(data[i]) 
    return ans



def sort_rt(a,b):
    d = {}
    sorted_rt = []
    followup_rt1=[]
    
    for i,j in zip(a,b):
        d[i] = j
    
    for key, value in sorted(d.items()):
        #print(key,value)
        sorted_rt.append(key)
        followup_rt1.append(value)
    return (sorted_rt,followup_rt1)




def normal_distribution(data):
    #data = df['Purchase']
    Cleaned_data = np.array(remove_outlier(data))
    mu = np.mean(Cleaned_data) 
    sigma = np.std(Cleaned_data, ddof=1)
    data_pdf = norm.pdf(Cleaned_data, loc = mu, scale = sigma)
    a,b=sort_rt(Cleaned_data,data_pdf)
    return a,b





```

```python
def apply_CI(sample_data,sample_size,total_sample):
    sample_CI = []
    for i in range(total_sample):
        sample = np.random.choice(sample_data, sample_size)
        a = norm.interval(alpha=0.10,loc=np.mean(sample_data),scale=stats.sem(sample_data))
        sample_CI.append(a)
    return sample_CI
        
```

```python
def remove_outlier(data):

    ans = []
    data_s = sorted(data)

    q1,q3 = np.percentile(data_s,[25,75])
    iqr = q1 - q3
    upper_bound = q3 - (1.5*iqr)
    lower_bound = q1 + (1.5*iqr)

    for i in range(len(data)):
        
        if data[i] > lower_bound and data[i] < upper_bound:
            ans.append(data[i]) 
            

    return ans
```

```python
def sort_rt(a,b):
    d = {}
    sorted_rt = []
    followup_rt1=[]
    
    for i,j in zip(a,b):
        d[i] = j
    
    for key, value in sorted(d.items()):
        #print(key,value)
        sorted_rt.append(key)
        followup_rt1.append(value)
    return (sorted_rt,followup_rt1)
```

```python
def normal_distribution(data):
    data = df['Age']
    Cleaned_data = np.array(remove_outlier(data))
    mu = np.mean(Cleaned_data) 
    sigma = np.std(Cleaned_data, ddof=1)
    data_pdf = norm.pdf(Cleaned_data, loc = mu, scale = sigma)
    a,b=sort_rt(Cleaned_data,data_pdf)
    return a,b



a = df['Age']
c,d=normal_distribution(a)
plt.title('Cleaned Normal Distribution for Age')
plt.plot(c,d)
plt.show()
```
